```python
import numpy as np
from scipy.optimize import bisect
import matplotlib.pyplot as plt
import ipywidgets as widgets

def solve_diff_eq(C0, C_inf, v, dp, Dl, t_max, dt, dx, L):
    k = v * dp / Dl
    Nt = int(t_max / dt)
    Nx = int(L / dx) + 1
    C = np.zeros((Nt + 1, Nx + 1))
    C[0, :] = C0
    for n in range(Nt):
        for i in range(1, Nx):
            C[n + 1, i] = C[n, i] + k * dt / dx**2 * (C[n, i + 1] - 2 * C[n, i] + C[n, i - 1])
    C[:, -1] = C_inf
    x = np.linspace(0, L, Nx + 1)
    return C, x



def solve_reactor_length(v, dp, Dl, C_inf, L, t_max, dt, dx):
    C, x = solve_diff_eq(C0, C_inf, v, dp, Dl, t_max, dt, dx, L)
    C_eff = C_inf * 0.99

    def residual(L):
        C, x = solve_diff_eq(C0, C_inf, v, dp, Dl, t_max, dt, dx, L)
        return C[-1, -1] - C_eff

    L_opt = bisect(residual, 0, L)
    return L_opt




# plot of solution


def update_plot(v, dp, Dl):
    L_opt = solve_reactor_length(v, dp, Dl, C_inf, L, t_max, dt, dx)
    C, x = solve_diff_eq(C0, C_inf, v, dp, Dl, t_max, dt, dx, L_opt)

    C_eff = 0.99 * C_inf  # 99% saturation
    idx = np.argwhere(C[-1, :] >= C_eff)[0][0]
    x_99 = x[idx]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.plot(x, C[-1, :])
    ax1.axhline(y=C_eff, color='r', linestyle='--')
    ax1.axvline(x=x_99, color='r', linestyle='--')
    ax1.set_xlabel('Position [m]')
    ax1.set_ylabel('Concentration [mol/m^3]')
    ax1.set_title('Percent Saturation')



# Plotting the contact time
    x = np.linspace(0, L[-1], 1000)
    tau = x/v
    ax2.plot(x, tau)
    ax2.set_xlabel('Reactor Length (x)')
    ax2.set_ylabel('Contact Time (tau)')
    ax2.set_title('Contact Time vs. Reactor Length')

    plt.show()

def reset_sliders(b):
    v.value = 0.1
    dp.value = 1
    Dl.value = 1e-9

v = widgets.FloatText(description='v', value=0.1)
dp = widgets.FloatText(description='dp', value=1)
Dl = widgets.FloatText(description='Dl', value=1e-9)

v_slider = widgets.FloatSlider(min=0.1, max=1, step=0.1, value=0.1, description='')
dp_slider = widgets.FloatSlider(min=0.1, max=10, step=0.1, value=1, description='')
Dl_slider = widgets.FloatSlider(min=1e-9, max=1e-7, step=1e-9, value=1e-9, description='')
reset_button = widgets.Button(description='Reset Sliders')
reset_button.on_click(reset_sliders)

C0 = 0  # Initial concentration
C_inf = 1  # Solute concentration in the feed
L = 1  # Reactor length
t_max = 10  # Maximum simulation time
dt = 0.1  # Time step
dx = 0.01  # Space step

widgets.interactive(update_plot, v=[0.1, 1, 0.1], dp=[0.1, 1, 0.1],
                    Dl=[0.1, 1, 0.1])
display(reset_button)



def main():
    C0=0
    C_inf=0.89
    v=0.1
    dp=1
    Dl=1
    t_max=1
    dt=0.1
    dx=0.01
    L=1

    update_plot(v, dp, Dl)

if __name__=="__main__":
    main()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import ipywidgets as widgets

# Constants
kf = 0.1
Ceff = 0.99
L = 1

# Calculating mass transfer coefficient
def mass_transfer_coefficient(v, dp, Dl):
    return kf*(v/Dl)/(1 + (v/Dl)*(dp/kf))

# Solving the differential equation
def solve_diff_eq(v, dp, Dl):
    k = mass_transfer_coefficient(v, dp, Dl)
    C0 = [0]
    x = np.linspace(0, L, 1000)
    sol = solve_ivp(lambda x, C: k*(Ceff - C), [0, L], C0, t_eval=x)
    return sol.t, sol.y[0]

def update_plot(v, dp, Dl):
    x, C = solve_diff_eq(v, dp, Dl)
    plt.figure(figsize=(12, 8))
    plt.subplot(121)
    plt.plot(x, C)
    plt.ylabel("Depth (L)")
    plt.xlabel("Concentration (C)")
    plt.title("% Saturation vs Depth")
    
    plt.subplot(122)
    v = np.linspace(0, v, 1000)
    t = [L/v for v in v]
    plt.plot(x, t)
    plt.xlabel("Depth (L)")
    plt.ylabel("Residence Time (t)")
    plt.title("Residence Time vs Depth")
    
    plt.tight_layout()
    plt.show()

v = widgets.FloatSlider(value=0.1, min=0.1, max=1.0, description='v')
dp = widgets.FloatSlider(value=0.1, min=0.1, max=1.0, description='dp')
Dl = widgets.FloatSlider(value=0.1, min=0.1, max=1.0, description='Dl')
widgets.interactive(update_plot, v=v, dp=dp, Dl=Dl)
```
